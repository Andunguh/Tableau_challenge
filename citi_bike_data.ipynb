```python
# Import Dependencies
import pandas as pd
```

```python
january_data = "Resources/January2021.csv"
february_data = "Resources/February2021.csv"

```

```python
january_df = pd.read_csv(january_data)
february_df = pd.read_csv(february_data)
```

```python
# print january_df
january_df.head()
```

```python
# print february_df
february_df.head()
```

```python
# Combine the data into a single dataset.  
citi_bike_df = pd.merge(january_df, february_df, how="left")
citi_bike_df.head()
```

```python
# remove index
citi_bike_df.style.hide_index()
```

```python
#convert df to csv
citi_bike_df.to_csv("Resources/citi_bikeData.csv", index=False)
```
